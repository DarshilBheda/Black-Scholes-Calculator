```python
# Imports
from wallstreet import Stock, Call, Put
from pandas_datareader import data as pdr
from yahoo_fin import stock_info as si
from yahoo_fin import options as oi
from pandas import ExcelWriter
import yfinance as yf
import pandas as pd
import datetime
import time
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
yf.pdr_override()
```

```python
# 2. Create List of Tickers to Analyze

# Pull all tickers currently on S&P 500
tickers = si.tickers_sp500()
tickers = [item.replace(".", "-") for item in tickers] # Yahoo Finance uses dashes instead of dots
```

```python
# 3. Concatenate list of all available call options for the nearest expiry date
df = pd.DataFrame(columns=['Contract Name', "Last Trade Date", "Strike", "Last Price", "Bid", "Ask", "Change", "% Change", "Volume","Open Interest","Implied Volatility"])

for ticker in tickers:
    try: 
        df = pd.concat([df, oi.get_calls(ticker)])
        time.sleep(2)
    except Exception:
        print(f"Could not gather data on {ticker}. Making a second attempt.")
        time.sleep(60)
        try:
            df = pd.concat([df, oi.get_calls(ticker)])
        except Exception as e:
            print (e)
            print(f"Could not gather data on {ticker}.")
```
